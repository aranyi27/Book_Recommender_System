## POPULARITY BASED RECOMMENDATION SYSTEM

```python
ratings_with_name = ratings.merge(books, on = 'ISBN')
```

```python
ratings_with_name
```

```python
ratings_with_name.groupby('Book-Title').count()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
```

```python
num_rating_df
```

```python
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title', as_index=False)['Book-Rating'].mean()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df[popular_df['num_ratings'] >= 250]
```

```python
popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending=False)
```

```python
popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

## COLLABORATIVE FILTERING BASED RECOMMENDER SYSTEM

```python
x= ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
x[x] #Boolean indexing to get true values
regular_users = x[x].index
```

```python
ratings_with_name
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(regular_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.shape
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
#%pip install scikit-learn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:6]

    for i in similar_items:
        print([pt.index[i[0]]])     
```

```python
recommend('The Notebook')
```
